# About the dataset ([source](https://www.kaggle.com/danielpanizzo/wine-quality))

Description of attributes (physicochemical) - inputs:
1. fixed acidity: most acids involved with wine or fixed or nonvolatile (do not evaporate readily)
2. volatile acidity: the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste
3. citric acid: found in small quantities, citric acid can add 'freshness' and flavor to wines
4. residual sugar: the amount of sugar remaining after fermentation stops, it's rare to find wines with less than 1 gram/liter and wines with greater than 45 grams/liter are considered sweet
5. chlorides: the amount of salt in the wine
6. free sulfur dioxide: the free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion; it prevents microbial growth and the oxidation of wine
7. total sulfur dioxide: amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine, but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine
8. density: the density of water is close to that of water depending on the percent alcohol and sugar content
9. pH: describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 on the pH scale
10. sulphates: a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant
11. alcohol: the percent alcohol content of the wine

Output variable:
- based on sensory data - median of at least 3 evaluations made by wine experts
- each expert graded the wine quality between 0 (very bad) and 10 (very excellent)

12. quality (score between 0 and 10)

In the data description, there is stated that the support vector machine model achieved the
best results. I will therefore want to prepare the data for this model. The model will predict whether a wine is bad (0-5) or good (5-10).

First, let's get an intuition of what the accuracy on the raw data is.

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import matthews_corrcoef

target = "quality"

# given a dataframe, create train test split and get SVC score
def print_score(dataframe, scale=False):
    dataframe = dataframe.copy()
    # use binning to split wines into good ones and bad ones, use 0 for bad and 1 for good
    dataframe['quality'] = pd.cut(dataframe['quality'], bins=[0, 5, 10], labels=[0, 1])

    X, y = get_X_and_y(dataframe)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    
    if scale:
        scaler = StandardScaler()
        scaler.fit(X_train)
        X_train = scaler.transform(X_train)
        X_test = scaler.transform(X_test)

    # default model
    model = SVC()
    print("Default model:")
    model.fit(X_train, y_train)
    print_metrics(y_test, model.predict(X_test))
    
    # find best parameters for our SVC model
    param = {
        'C': [0.1, 0.5, 1],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    }
    grid_svc = GridSearchCV(model, param_grid=param, scoring='accuracy', cv=10)
    grid_svc.fit(X_train, y_train)
    
    # using grid model itself
    print("Model with 'optimal' parameters:")
    print_metrics(y_test, grid_svc.predict(X_test))
    
    # using its best estimator
    # print_metrics(y_test, grid_svc.best_estimator_.predict(X_test))
    
    # using new model with found best parameters
    """best_params = grid_svc.best_params_
    model = SVC(best_params)
    model.fit(X_train, y_train)
    print_metrics(y_test, model.predict(X_test))"""
    
# split dataframe into dataframe without target and target dataframe
def get_X_and_y(data, target=target):
    X = data.drop(target, axis=1)
    y = data[target]
    
    return X, y

def print_metrics(y_test, y_predicted):
    print(f"Accuracy score: {accuracy_score(y_test, y_predicted)}")
    print(f"Confusion matrix:\n {confusion_matrix(y_test, y_predicted)}")
    print(f"Classification report:\n {classification_report(y_test, y_predicted)}")
    print(f"Matthews corr coef:\n {matthews_corrcoef(y_test, y_predicted)}")

print_score(df)
```

Let's apply standard scaling, i. e. substract the mean and divide by standard deviation.

```python
print_score(df, scale=True)
```

We can see a 12% increase in accuracy!

Let's find some correlations. Based on the description of columns, my assumptions are:
1. negative correlation of **volatile acidity** and **quality**
2. **total sulfur dioxide** correlates with **free sulfur dioxide** -> dependant attributes
3. **density** depends on the percent **alcohol** and sugar content (i. e. **residual sugar**)
4. **sulphates** can contribute to **free sulfur dioxide** and **total sulfur dioxide**

```python
candidates = [("volatile acidity", "quality"),
 ("total sulfur dioxide", "free sulfur dioxide"),
 ("density", "alcohol"),
 ("density", "residual sugar"),
 ("sulphates", "free sulfur dioxide"),
 ("sulphates", "total sulfur dioxide")]

for a, b in candidates:
    print(f"{a:>20} x {b:20} {df[a].corr(df[b])}")
```

All assumptions seem to hold, apart from the fourth.

Now let's print the whole matrix and find other correlations.

```python
import seaborn as sn

corr_matrix = df.corr()
corr = corr_matrix.stack()

# remove correlation with itself
corr = corr[corr.index.get_level_values(0) != corr.index.get_level_values(1)]

# print only significant correlations
print(corr[corr > 0.5])

# make figure bigger
plt.subplots(figsize=(10,10))

sn.heatmap(corr_matrix, annot=True)

plt.show()
```

```python
# show how quality depends on some attribute
def quality_corr(attr):
    fig = plt.figure(figsize = (10,6))
    sn.barplot(x='quality', y=attr, data=df)

quality_corr("citric acid")
```

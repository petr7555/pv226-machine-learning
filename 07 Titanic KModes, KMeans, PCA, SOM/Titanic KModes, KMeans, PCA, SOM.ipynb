```python
import pandas as pd

df = pd.read_csv("./data/train.csv").set_index("PassengerId")

df.head()
```

```python
X = df.loc[:,df.columns != "Survived"]
y = df["Survived"]
```

```python
X.info()
```

I can see, there are some columns containing NaN values.
These can't be be handled by KModes, so I need to fill those missing values.

(warning "A value is trying to be set on a copy of a slice from a DataFrame.
Try using .loc[row_indexer,col_indexer] = value instead" can be ignored)

```python
X.loc[:,'Cabin'] = X.loc[:,'Cabin'].fillna("")
X.loc[:,'Embarked'] = X.loc[:,'Embarked'].fillna("")
X.loc[:,'Age'] = X.loc[:,'Age'].fillna(X["Age"].mean())
```

```python
print(X.info())
print(X.isna().any())
```

I can see, that no columns contain NaN now.

```python
from kmodes.kmodes import KModes

km = KModes(n_clusters=3, init='Huang', n_init=5, verbose=1) # or "Cao"

y_km = km.fit_predict(X)

# Print the cluster centroids
print(km.cluster_centroids_)
```

```python
# concatenate cluster outputs with data
y_km_series = pd.Series(y_km, name="Cluster")
with_clusters = pd.concat([X,y,y_km_series], axis=1, join="inner")
with_clusters
```

```python
# export to csv
with_clusters.to_csv("with_clusters.csv")
# you can open it in any spreadsheet app (Excel, GoogleSpreadsheets,...)
```

```python
# let's try if I can see some difference among clusters
with_clusters.groupby("Cluster").mean()
```

## Choosing K by comparing Cost against each K
https://www.kaggle.com/ashydv/bank-customer-clustering-k-modes-clustering

```python
import numpy as np
from matplotlib import pyplot as plt

cost = []
for num_clusters in list(range(1,5)):
    kmodes = KModes(n_clusters=num_clusters, init = "Cao", n_init = 1, verbose=1)
    kmodes.fit_predict(X)
    cost.append(kmode.cost_)

y_axis = np.array([i for i in range(1,5,1)])
plt.plot(y_axis, cost)
```

```python
import seaborn as sns

kmodes = KModes(n_clusters=2, init = "Cao", n_init = 1, verbose=1)
y_km = kmodes.fit_predict(X)
    
X_combined = X.copy()
X_combined["Cluster"]=y_km
cluster_0 = X_combined[X_combined['Cluster'] == 0]
cluster_1 = X_combined[X_combined['Cluster'] == 1]

# visualizations like this one can help in identification of clusters
plt.subplots(figsize = (15,5))
sns.countplot(x=X_combined['Embarked'],
              order=X_combined['Embarked'].value_counts().index,
              hue=X_combined['Cluster'])
plt.show()
```

# Self-organizing map

SOM can be used both as way of mapping categorical data to numerical, 
or to reduce dimensionality for subsequent clustering 
(related to first approach).
Autoencoders reduce dimensionality as well. 
Let's use it to cluster Titanic passengers.

- [Simple explanation of SOMs](https://www.kaggle.com/residentmario/self-organizing-maps)
- [SimpSOM](https://simpsom.readthedocs.io/en/master/index.html)
- [minisom](https://github.com/JustGlowing/minisom)

It needs only numerical values, so I will drop "Ticket" column, which is string.

```python
X = X.drop(columns=["Ticket"])
X.info()
```

```python
# following this tutorial 
# https://medium.com/@dalicodes/overview-of-self-organizing-maps-som-with-its-python-implementation-in-determining-safe-airlines-db8f6018a2b

import SimpSOM as sps

net = sps.somNet(20, 20, X.values, PBC=True)
net.train(0.01, 20000)
#net.save(‘filename_weights’)
net.nodes_graph(colnum=0)
```

```python
net.diff_graph()
```

```python
import matplotlib.pyplot as plt

# Project the datapoints on the new 2D network map
prj = np.array(net.project(X.values))
plt.scatter(prj.T[0],prj.T[1])
plt.show()

prj.T.shape # (2, 891)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit_predict(prj)
X_with_clusters = X.copy()
X_with_clusters["Cluster"]=kmeans
X_with_clusters.groupby("Cluster").mean()
```

We now have two clusters of data, that have been clustered 
on projection of SOM with reduced dimensionality (not really, I still have 20x20 weights).

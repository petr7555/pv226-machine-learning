```python
import pandas as pd

# load csv data
df = pd.read_csv("./data/wines.csv")

df.head()
```

```python
df.info()
```

```python
red_df = df[df["type"]=="red"]
red_df_without_type = red_df.drop("type", axis="columns")
df = red_df_without_type
df.describe()
```

```python
from sklearn.model_selection import train_test_split

target = "quality"
X = df.drop(target, axis=1)
y = df[target]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Let's start with regression task.

```python
# https://autokeras.com/tutorial/structured_data_classification/
import autokeras as ak

# Initialize the structured data classifier.
clf = ak.StructuredDataClassifier(
    overwrite=False, # When building of model fails, 
                     # set to False and then back to True
    max_trials=5) # It tries 5 different models.

# Feed the structured data classifier with training data.
clf.fit(x_train,
        y_train,
        # Split the training data and use the last 15% as validation data.
        validation_split=0.15,
        epochs=50)

# returns scalar test loss (if the model has a single output and no metrics)
# or list of scalars (if the model has multiple outputs and/or metrics)
print(clf.evaluate(x_test, y_test))
```

Let's turn this into classification task: bad / good wine.

```python
# use binning to split wines into good ones and bad ones, use 0 for bad and 1 for good
y = pd.cut(y, bins=[0, 5, 10], labels=[0, 1])
y = y.astype("int64")
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Initialize the structured data classifier.
clf = ak.StructuredDataClassifier(
    overwrite=False, # When building of model fails, 
                     # set to False and then back to True
    max_trials=5) # It tries 5 different models.

# Feed the structured data classifier with training data.
clf.fit(x_train,
        y_train,
        # Split the training data and use the last 15% as validation data.
        validation_split=0.15,
        epochs=50)

# returns scalar test loss (if the model has a single output and no metrics)
# or list of scalars (if the model has multiple outputs and/or metrics)
print(clf.evaluate(x_test, y_test))
```
